```python
import warnings
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
warnings.filterwarnings('ignore')# evitar os pop-ups de warning já previstos
```

```python
from pycaret.datasets import get_data
df = get_data('credit')
```

```python
df = df.drop("id", axis=1)
```

```python
le = LabelEncoder()
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)

# create a DataFrame to hold the encoded columns
encoded_cols = pd.DataFrame()

# loop through each column in the DataFrame
for col in df.columns:
    if df[col].dtype == 'object':  # check if column is categorical
        # use label encoding for categorical columns with 2 or less unique values
        if len(df[col].unique()) <= 2:
            df[col] = le.fit_transform(df[col])
        # use one-hot encoding for categorical columns with more than 2 unique values
        else:
            encoded_array = ohe.fit_transform(df[[col]])
            encoded_df = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out([col]))
            encoded_cols = pd.concat([encoded_cols, encoded_df], axis=1)
    else:  # column is numerical, so no encoding is needed
        encoded_cols = pd.concat([encoded_cols, df[[col]]], axis=1)

# concatenate the encoded columns with the original DataFrame
df_encoded = pd.concat([encoded_cols, df.select_dtypes(include=['float64', 'int64'])], axis=1)

# your DataFrame is now encoded and ready to be used in a machine learning model
```
